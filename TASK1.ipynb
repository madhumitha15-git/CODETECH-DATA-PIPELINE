```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
```

```python
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
```

```python
df
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.drop('date_added', axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['type'] = encoder.fit_transform(df['type'])
df['rating'] = encoder.fit_transform(df['rating'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['release_year','year_added','month_added']] = scaler.fit_transform(
    df[['release_year','year_added','month_added']]
)
```

```python
df
```

```python
df['duration_value'] = df['duration'].str.extract(r'(\d+)')
```

```python
df['duration_value'] = df['duration_value'].astype(int)
```

```python
df
```

```python
df['duration_type'] = df['duration'].apply(
    lambda x: 'Minutes' if 'min' in x else 'Seasons'
)
```

```python
df
```

```python
df.drop('duration', axis=1, inplace=True)
```

```python
df
```

```python
df['description'] = df['description'].str.lower()
```

```python
import re

df['description'] = df['description'].apply(
    lambda x: re.sub(r'[^a-zA-Z\s]', '', x)
)
```

```python
df['description'] = df['description'].str.strip()
```

```python
df
```

```python
df.to_csv("netflix_processed.csv",index=False)
```
